```python
# Write a Python function that accepts a list of numbers and returns the largest number

def largestNumber(numbers):
    largest=numbers[0]
    for i in numbers:
        if i>largest:
            largest=i
    return largest

numbers=[25, 48, 12, 67, 39]
result = largestNumber(numbers)
print(result)

```

```python
# Write a Python function that accepts a list of numbers and returns the smallest number

def smallestNumber(numbers):
    smallest=numbers[0]
    for i in numbers:
        if i<smallest:
            smallest=i
    return smallest
numbers= [25, 48, 12, 67, 39]
result= smallestNumber(numbers)
print(result)
    
```

```python
# Write a Python function to sum of all numbers in a list

def sumOfNumbers(numbers):
    total=0
    for i in numbers:
        total=total+i
    return total
numbers=[10, 20, 30, 40]
result=sumOfNumbers(numbers)
print(result)
    
```

```python
# Write a function that accepts a list and returns the count of even numbers.

def countOfEven(numbers):
    count=0
    for i in numbers:
        if i%2==0:
            count=count+1
    return count
numbers=[10, 15, 20, 35, 40]
print(countOfEven(numbers))
```

```python
#write a function that accepts a list of numbers and returns the average

def listAvg(numbers):
    count=0
    total=0
    for i in numbers:
        count=count+1
        total+=i
    nosAvg=total/count
    return nosAvg
numbers=[80, 75, 90, 85, 70]
print(listAvg(numbers))

```

```python
#write a function to find a prime number

def primeNumber(a):
    
    prime=True
    if a<=1:
        prime=False
    for i in range(2,a//2+1):
        if a%i==0:
            prime=False
    return prime
print(primeNumber(17))
```

```python
#write a function to find a palindrome number

def palindrome(a):
    original= a
    num=a
    reversed= 0
    while num>0:
        lastDigit=num%10
        reversed=reversed*10+lastDigit
        num=num//10
    if original==reversed:
        return True
    else:
        return False      
    
print(palindrome(1234))
```

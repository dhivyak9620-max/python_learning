```python
#Task 1: Search for an Item in a Tuple 

fruits = ("Apple", "Banana", "Orange", "Mango") 

if "Orange" in fruits:
    print("Orange Found ")
else:
    print("Orange Not Found")

if "Cherry" in fruits:
    print("Cherry Found ")
else:
    print("Cherry Not Found")
```

```python
#Task 2: Count the Occurrence of an Element

numbers = (10, 20, 10, 30, 10, 40)
print("10 appears", numbers.count(10), "times")
```

```python
#Task 3: Print Tuple Elements Using a While Loop

colors = ("Red", "Green", "Blue", "Yellow")

i=0
while i<len(colors):
    print(colors[i])
    i+=1
```

```python
#Task 4: Calculate the Sum and Average
marks = (80, 75, 90, 85, 70)
total=0

for i in marks:
    total=total+i
avg=total/len(marks)
print("sum =",total)
print("avg =",avg)

# sum by using a keyword:
totalSum = sum(marks)
print("totalSum =", totalSum)



```

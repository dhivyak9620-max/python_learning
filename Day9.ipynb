```python
#Write a Python function to count the number of digits in a given number using a while loop.

def countNumbers(num):
    count=0
    while num>0:
        num=num//10
        count+=1
    return count
num=12345
countNumbers(num)
```

```python
#Write a function that returns the sum of all digits in a number.

def sumOfDigits(num):
    total=0
    while num>0:
        total=total+num % 10
        num=num//10
    return total
num=12345
sumOfDigits(num)
```

```python
#write a function to findout whether the number is an armstrong number:

def armstrongNumber(num):
    copyOfNum=num
    secondCopy=num
    count=0
    total=0
    while num>0:
        num=num//10
        count+=1
    while copyOfNum>0:
        remainder=copyOfNum % 10
        total=total + remainder**count
        copyOfNum=copyOfNum // 10

    if secondCopy==total:
        return True
    else:
        return False
num=153
armstrongNumber(num)
```

```python
#Write a Python function that returns the factorial of a given positive integer.

def factorialOfNum(num):
    factorial=1
    for i in range(num,1,-1):
        factorial=i*factorial
    return factorial
num=4
print(factorialOfNum(num))

```

```python
#Problem 1: Student Result

name=input("Enter your Name:")
mathMarks=int(input("Enter your Math subject marks:"))
scienceMarks=int(input("Enter your Science subject marks:"))
socialStudiesMarks=int(input("Enter your Social Studies subject marks:"))

total=mathMarks+scienceMarks+socialStudiesMarks
average=total//3
print("Name:",name)
print("Total:",total)
print("Average:",average)

```

```python
#Problem 2: Student Grade calculator:
name=input("Enter your Name:")
mathMarks=int(input("Enter your Math subject marks:"))
scienceMarks=int(input("Enter your Science subject marks:"))
socialStudiesMarks=int(input("Enter your Social Studies subject marks:"))

total=mathMarks+scienceMarks+socialStudiesMarks
average=round(total/3,2)
if average>=90:
    grade="Grade A"
elif average>=75:
    grade="Grade B"
elif average>=50:
    grade="Grade C"
else:
    grade="Fail"

print("Name:",name)
print("Total:",total)
print("Average:",average)
print("Grade",grade)
```

```python
#Problem 3: Biggest Among Three Numbers

numbers=[]

for i in range(3):
    userInput=int(input("Enter any number:"))
    numbers.append(userInput)
    
largestNumber=numbers[0]
for i in numbers:
    if i>largestNumber:
        largestNumber=i
    else:
        largestNumber=largestNumber

print("Largest Number is",largestNumber)


    
```

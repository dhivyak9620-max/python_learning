```python
#Task 1: Safe Calculator
 
def safe_division():
    try:
        num1=int(input("Enter any number:"))
        num2=int(input("Enter any number:"))
        result=(num1/num2)
        if result>0:
            print("Positive")
        else:
            print("negative")
    except ZeroDivisionError:
        print("Cannot divided by zero")
    except ValueError:
        print("Non Numeric Input")
    finally:
        print("Calculation Completed")
safe_division()


    
```

```python
#Task 2: Student Age Verification 

def check_age():
    i=0
    while i==0:
        try:
            age=int(input("Enter Your Age:"))
            if age<18:
                print("Minor")
            else:
                print("Adult")
            i+=1
        except ValueError:
            print("Invalid Input")
        finally:
            print ("Age verification finished")
check_age()

        
```

```python
# Task 3: Number List Average 

def find_average():
    try:
        a=int(input("How many numbers you want to add in a list:"))
        numList=[]
        for i in range(a):
            num=int(input(f"Enter the number{i+1}" ))
            numList.append(num)
        listAverage=sum(numList)/a
        if listAverage>50:
            print("Avg is greater than 50")
        else:
            print("Avg is less than 50")
    except ValueError:
        print("Invalid Input")
    except ZeroDivisionError:
        print("Cannot be divided by Zero")
    finally:
        print("Average calculation completed")
find_average()
    

```

```python
#Task 4: Password Login System

def login():
    i=1
    while i<=3:
        try:
            userPassword=int(input(f"Attempt{i}:Enter your login password:"))
            if userPassword==1234:
                print("Login Successful")
            else:
                print("Invalid Password")
            i+=1
        except ValueError:
            print("Invalid Input")
            i+=1
        finally:
            print("Login Process Ended")
login()
```

```python
#Task 5: Shopping Bill Calculator 

def shopping_bill():
    try:
        purchasedItemsCount=int(input("How many items have you purchased:"))
        purchasedItemsList=[]
        for i in range(purchasedItemsCount):
            purchasedItems=int(input(f"Item No:{i+1}"))
            purchasedItemsList.append(purchasedItems)
        Total=sum(purchasedItemsList)
        billSplitCount=int(input("with how many people you want to share the bill:"))
        billSplit=Total/billSplitCount
        if billSplit>1000:
            print("High Bill")
        else:
            print("Affordable Bill")
    
    except ValueError:
        print("Invalid Input")
    except ZeroDivisionError:
        print("Cannot divided by zero")
    finally:
        print("Billing Process Completed")
shopping_bill()
```

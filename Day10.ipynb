```python
#write a function to find a perfect number
def perfectNumber(num):
    original=num
    total=0
    for i in range(1,num-1):
        if num%i==0:
            total=total+i
    return original==total
num=12
print(perfectNumber(num))
```

```python
def countFrequency(num,digit):
    count=0
    while num>0:
        lastdigit=num%10
        if lastdigit==digit:
            count+=1
        num=num//10
    return count  

num=int(input("Enter any number"))
digit=int(input("Enter the digit to check frequency in the given number"))
print(countFrequency(num,digit))
```

```python
#write the function to convert the decimal to binary
def decimalToBinary(num):
    quotient=num
    digitsForbinary=0
    binary=0
    while quotient>0:
        remainder=quotient%2
        digitsForbinary=digitsForbinary*10+remainder
        quotient=quotient//2
    
    while digitsForbinary>0:
        lastdigit=digitsForbinary%10
        binary=binary*10+lastdigit
        digitsForbinary=digitsForbinary//10
    
    return binary

num=int(input("Enter any number"))
print(decimalToBinary(num))
```

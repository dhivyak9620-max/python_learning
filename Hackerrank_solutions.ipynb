```python
#Task_1: If else

'''Write a Python program that determines whether a number is "Weird" or "Not Weird" 
based on whether it is odd or even and the specified number ranges.'''
n=int(input())
if n%2==0:
    if n>=2 and n<=5:
        print("Not Weird")
    elif n>=6 and n<=20:
        print("Weird")
    elif n>20:
        print("Not Weird")
else:
    print("Weird")
```

```python
#Task 2: Arithmetic operators

'''Write a Python program to perform addition, subtraction, and multiplication 
on two input integers and display the results.'''

a=int(input())
b=int(input())
total=a+b
difference=a-b
product=a*b
print(total)
print(difference)
print(product)
```

```python
#Task 4: Print Squares

'''Write a Python program to print the square of every non-negative integer 
less than a given number.'''

n=int(input())
for i in range(n):
    print(i*i)
```

```python
#Task 5: Leap Year Checker

'''Write a function that checks whether a given year is a 
leap year by following the Gregorian calendar rules.'''

def is_leap(year):
    leap = False
    
    if year%4 == 0:
        leap = True
    if year%100 == 0:
        leap = False
    if year%400 == 0:
        leap = True
        
    return leap
```

```python
#Task 6:Print Consecutive Numbers

'''Write a Python program to print consecutive numbers from 1 to n on a single line 
without using string methods.'''

n=int(input())
for i in range(1,n+1):
    print(i,end="")
```

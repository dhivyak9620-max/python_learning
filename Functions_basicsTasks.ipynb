```python
#create a function to print even or odd:
def evenOdd(a):
    if a%2==0:
        print(a,"is a even number")
    else:
        print(a,"is a odd number")
evenOdd(3)
evenOdd(14)
```

```python
#create a function to print table:

def table(a):
    for i in range(1,11):
        print(a,"x",i,"=",a*i)
table(5)
```

```python
#create the function to return largest value

def largestNo(a,b,c):
    if a>b and a>c:
        return (a, "is largest")
    elif b>a and b>c:
        return (b, "is largest")
    else:
        return (c, "is largest")
result=largestNo(10,25,15)
print(result)
```
